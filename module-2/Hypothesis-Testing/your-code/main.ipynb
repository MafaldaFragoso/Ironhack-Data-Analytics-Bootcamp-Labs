# Challenge 2 - Hypothesis Tests

In this section of the lab, we will test whether the hourly wage of all hourly workers is significantly different from $30/hr. Import the correct one sample test function from scipy and perform the hypothesis test for a 95% two sided confidence interval.

```python
from scipy.stats import ttest_1samp

hourly = employees["Hourly Rate"].loc[employees["Salary or Hourly"]=="Hourly"]
H0 = 30
H1 != 30

ttest,pval = ttest_1samp(hourly,H0)

print("ttest",ttest)
print("p-value",pval)

if pval <0.05:
    print("we reject null hypothesis")
else:
    print("we accept null hypothesis")
```

We are also curious about salaries in the police force. The chief of police in Chicago claimed in a press briefing that salaries this year are higher than last year's mean of $86000/year a year for all salaried employees. Test this one sided hypothesis using a 95% confidence interval.

Hint: A one tailed test has a p-value that is half of the two tailed p-value. If our hypothesis is greater than, then to reject, the test statistic must also be positive.

```python
H0 = 86000
police_salaries = employees[((employees['Department'] == 'POLICE') & (employees['Salary or Hourly'] == 'Salary'))]

ttest,pval = ttest_1samp(police_salaries['Annual Salary'], H0)

print("ttest",ttest)
print("p-value",pval/2)

if pval/2 <0.05:
    print("we reject null hypothesis")
else:
    print("we accept null hypothesis")
```

Using the `crosstab` function, find the department that has the most hourly workers. 

```python
department = pd.crosstab(employees['Department'],employees['Salary or Hourly']).sort_values(by='Hourly',ascending=False).reset_index()
department
```

The workers from the department with the most hourly workers have complained that their hourly wage is less than $35/hour. Using a one sample t-test, test this one-sided hypothesis at the 95% confidence level.

```python
# H0: workers from STREETS & SAN department >= 35 hourly wage
# H1: workers from STREETS & SAN department < 35 hourly wage

department_name = department.values[0][0]
department_salaries = employees[(employees['Department'] == department_name) & (employees['Salary or Hourly'] == 'Hourly')]
department_salaries

statistics = stats.ttest_1samp(department_salaries['Hourly Rate'], 35)
print("P Value:", statistics[1])
print("Statistics:", statistics[0])
```
